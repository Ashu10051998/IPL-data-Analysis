```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
teams_df = pd.read_csv('/content/drive/MyDrive/1st capstone project data/teams.csv')

home_away_df = pd.read_csv('/content/drive/MyDrive/1st capstone project data/teamwise_home_and_away.csv')

average_df = pd.read_csv('/content/drive/MyDrive/1st capstone project data/most_runs_average_strikerate.csv')

players_df = pd.read_excel('/content/drive/MyDrive/1st capstone project data/Players.xlsx')
file_path="/content/drive/MyDrive/1st capstone project data/"
ipl_df=pd.read_csv(file_path +'deliveries.csv')
matches_file_path='/content/drive/MyDrive/1st capstone project data/'
ipl_match_df = pd.read_csv(matches_file_path +'matches.csv')
```

```python
# Performances of all the teams on their home conditions

plt.figure(figsize=(15,5))
home_condition_performance = home_away_df.groupby(['team'])['home_wins'].sum().sort_values(ascending=False)[:]

home_condition_performance.plot()
home_condition_performance.plot(kind = 'bar')

plt.title('Most sucessful teams at home venues')
plt.xlabel('Teams')
plt.ylabel('Most Wins')
#  Performances of all the teams on their away conditions


plt.figure(figsize=(15,5))
away_condition_performance = home_away_df.groupby(['team'])['away_wins'].sum().sort_values(ascending=False)[:]

away_condition_performance.plot()
away_condition_performance.plot(kind = 'bar')

plt.title('Most sucessful teams at away venues')
plt.xlabel('Teams')
plt.ylabel('Most Wins')
```

```python
# Highest overall win percentage 


home_away_df['overall_win_percentage'] = (home_away_df['home_win_percentage'] + home_away_df['away_win_percentage'])/2    #Adding new column in original data frame 
plt.figure(figsize=(8, 5))
data = home_away_df.groupby(['team'])['overall_win_percentage'].sum().sort_values(ascending=False)[:]
sns.barplot(y=data.index, x=data , orient='h')
plt.title('Highest Overall Win Percentage')
plt.xlabel('Win Percentage')
plt.ylabel('Teams')
```

```python
# Top 5 players with most runs

plt.figure(figsize=(7, 3))
data = average_df.groupby(['batsman'])['total_runs'].sum().sort_values(ascending=False)[:5]
sns.barplot(y=data, x=data.index , orient='v')
plt.title('Most runs in IPL History')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
# Top 5 players getting dismissed most number times

plt.figure(figsize=(10,4))
data = average_df.groupby(['batsman'])['out'].sum().sort_values(ascending=False)[:5]
sns.barplot(y=data.index,x=data,orient='h')
plt.title('Most Dismissals in IPL History')
plt.xlabel('Dismissals')
plt.ylabel('Batsman')
plt.show()
```

```python
# Players who faced most numbers of delivers

plt.figure(figsize=(10,4))
most_deliveries = average_df.groupby(['batsman'])['numberofballs'].sum().sort_values(ascending=False)[:5]

most_deliveries.plot()

plt.title('Most Number Of Deliveries Faced By Batsman')
plt.xlabel('Deliveries')
plt.ylabel('Batsman')

plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Top 5 players with highest Batting average

plt.figure(figsize=(8,5))
most_deliveries = average_df.groupby(['batsman'])['average'].sum().sort_values(ascending=False)[:5]

most_deliveries.plot(kind='bar')

plt.title('Best Overall Batting Average')
plt.xlabel('Batsman')
plt.ylabel('Batting Average')
```

```python
plt.figure(figsize=(8,5))
most_deliveries = average_df.groupby(['batsman'])['strikerate'].sum().sort_values(ascending=False)[:5]

most_deliveries.plot()

plt.title('Best Overall Batting Strike Rate')
plt.xlabel('Batsman')
plt.ylabel('Batting Strike Rate')
```

```python
#toss Decisions 

toss_decision_type =['Field','Bat']
toss_decison=list(ipl_match_df['toss_decision'].value_counts())
fig=plt.figure(figsize=(15,10.5))
plt.title('Toss decisions',fontsize=25)
plt.pie(toss_decison,labels=toss_decision_type,autopct='%1.2f%%')
```

```python
#Team decision of of batting or fielding after winning of toss effect on match result

#Winning percetage after choosing Field or Bat in toss decision
Counts_of_toss_decsion =list(ipl_match_df['toss_decision'].value_counts())
print(Counts_of_toss_decsion)
toss_decision=list(ipl_match_df['toss_decision'])
print(toss_decision)
team_winning_toss=list(ipl_match_df['toss_winner'])
print(team_winning_toss)
winner=list(ipl_match_df['winner'])
fielding_decision_count=0
print(winner)
i=0
batting_decision_count=0
j=0
while i<len(toss_decision):
 if toss_decision[i]=='field':
  if team_winning_toss[i]==winner[i]:
   i=i+1
   fielding_decision_count=fielding_decision_count+1
  else:
   i=i+1
 else:
  i=i+1
while j<len(toss_decision):
 if toss_decision[j]=='bat':
  if team_winning_toss[j]==winner[j]:
   j=j+1
   batting_decision_count=batting_decision_count+1
  else:
   j=j+1
 else:
  j=j+1

print(fielding_decision_count)
print(batting_decision_count)

plt.figure(figsize=(6,8))
Y_axis_data = [fielding_decision_count*100/Counts_of_toss_decsion[0],batting_decision_count*100/Counts_of_toss_decsion[1]]
x_axis_data =['field','Bat']
sns.barplot(y=Y_axis_data,x=x_axis_data,orient='v',)
plt.ylim([0,70])
plt.text(0,Y_axis_data[0],'55.93%')
plt.text(1,Y_axis_data[1],'45.77%')
```
